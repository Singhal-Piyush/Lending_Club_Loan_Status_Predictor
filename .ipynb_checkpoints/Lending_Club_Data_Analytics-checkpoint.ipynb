### a. Data Cleaning

In this section we will be performing cleaning and transformation operations on the dataset to make the data clean and easier to handle. 

```python
dataset_loan.isnull().sum()
```

As we have more than 100 columns in the dataframe dataset_loan it's very difficult to check one by one the number of null_values and evaluate which columns are necessary for our data analysis of Lending Club. Hence will be making a new dataframe which contains these information and will be know as dataset_meta along with the descriptive information about the columns. 

```python
# making a data_meta_function
def dataset_meta_function(dataset):
    global dataset_meta
    dataset_meta = pd.DataFrame(columns=["Column_Name","NaN_Values","NaN_Value_Percentage","Non_NaN_Values","Unique_Values","Dtype"])
    l = len(dataset)
    for i in dataset.columns:
        n=dataset[i].isnull().sum()
        dataset_meta.loc[len(dataset_meta),:]=[i,n,n/l*100,
        dataset_loan[i].notnull().sum(),
        dataset_loan[i].nunique(),
        dataset_loan[i].dtypes]
    dataset_meta = pd.merge(dataset_meta,dataset_meta_desc,how = "left",left_on = "Column_Name",right_on = "Column_Name")
#     dataset_meta = dataset_meta.set_index("Column_Name").join(dataset_meta_desc.set_index("LoanStatNew"),how="left")
# print("Selected Dataset has" +dataset.shape(1)+"columns. \n There are following "+ +"numbers of column with Null Values")
    return(dataset_meta)
```

```python
start_time=time.time()
meta_data_1= dataset_meta_function(dataset_loan)
print("%s seconds" %(time.time()-start_time))
```

## Reducing Memory Usage 

```python
dataset_loan.info(memory_usage="deep")
```

```python
data_dtype = dataset_loan.dtypes.unique()
print(data_dtype)
```

We have 105 columns with Float64, 4 columns with Int64 and 36 columns with Object data types, we can downcast them to appropiate data type according to the values present in the column to reduce the memory usage by the dataset.

Memory usage by each dtypes can be shown using below code. 

```python
data_dtype = dataset_loan.dtypes.unique()
memory_usage_total = 0
for i in data_dtype:
    memory_usage = dataset_loan.select_dtypes(include = i).memory_usage(deep=True).sum()
    memory_usage = memory_usage/1024**2
    memory_usage_total = memory_usage_total + memory_usage
    print("Memory usage for {} dtypes is {:.3f} MB".format(i,memory_usage))
    
print("Total  memory usage is {:.3f} MB".format(memory_usage_total))
```

We will be downgrading Float64 to Float16 and Float32 to reduce occupied memory and for faster operation on the data. Similarly Int64 to Int8,Int16 and Int32. Whereas we will convert Object data type to Categorical and DateTime Dtype. 

Showing the Memory Usage by each columns in the dataset

```python
(dataset_loan.memory_usage(deep=True))/1024**2
```

Please note that the above values are in MB.

##### Converting int64 to appropriate int dtypes

```python
for i in dataset_loan.select_dtypes(include="int64").columns:
    dataset_loan[i] = pd.to_numeric(dataset_loan[i],downcast="unsigned")
```

##### Converting float to appropriate float dtypes

```python
for i in dataset_loan.select_dtypes(include="float64").columns:
    dataset_loan[i] = pd.to_numeric(dataset_loan[i],downcast="float")
```

##### Converting Object to Datetimens[] and Categorical Dtypes

```python
start_time = time.time()
for i in dataset_loan.columns:
    if dataset_loan[i].dtypes == "object":
        try :
            dataset_loan[i] = pd.to_datetime(dataset_loan[i])
        except ValueError :
            try:
                dataset_loan[i] = dataset_loan[i].astype("category")
            except ValueError :
                pass
print("%s seconds" %(time.time()-start_time))
```

```python
data_dtype = dataset_loan.dtypes.unique()
memory_usage_total = 0
for i in data_dtype:
    memory_usage = dataset_loan.select_dtypes(include = i).memory_usage(deep=True).sum()
    memory_usage = memory_usage/1024**2
    memory_usage_total = memory_usage_total + memory_usage
    print("Memory usage for {} dtypes is {:.3f} MB".format(i,memory_usage))
    
print("Total  memory usage is {:.3f} MB".format(memory_usage_total))
```
